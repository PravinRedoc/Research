```python
def draw_line(coef,intercept, mi, ma):
    # for the separating hyper plane ax+by+c=0, the weights are [a, b] and the intercept is c
    # to draw the hyper plane we are creating two points
    # 1. ((b*min-c)/a, min) i.e ax+by+c=0 ==> ax = (-by-c) ==> x = (-by-c)/a here in place of y we are keeping the minimum value of y
    # 2. ((b*max-c)/a, max) i.e ax+by+c=0 ==> ax = (-by-c) ==> x = (-by-c)/a here in place of y we are keeping the maximum value of y
    points=np.array([[((-coef[1]*mi - intercept)/coef[0]), mi],[((-coef[1]*ma - intercept)/coef[0]), ma]])
    plt.plot(points[:,0], points[:,1])
   
```

```python
#Support Vector Machines
regulparam = [0.001, 1, 100]
for k in regulparam:
    plt.figure(figsize=(20,5))
    for j,i in enumerate(ratios):    
        plt.subplot(1, 4, j+1)
        X_p=np.random.normal(0,0.05,size=(i[0],2))
        X_n=np.random.normal(0.13,0.02,size=(i[1],2))
        y_p=np.array([1]*i[0]).reshape(-1,1)
        y_n=np.array([0]*i[1]).reshape(-1,1)
        X=np.vstack((X_p,X_n))
        y=np.vstack((y_p,y_n))
#         scalar = StandardScaler()
#         X=scalar.fit_transform(X)
        clf = SVC(C=k,kernel='linear')
        clf.fit(X, y)
        plt.scatter(X_p[:,0],X_p[:,1])
        plt.scatter(X_n[:,0],X_n[:,1],color='red')
        ma,mi= max(X[:,1]),min(X[:,1])
        intercept = clf.intercept_
        coef = clf.coef_[0]
        sv = clf.support_vectors_
        draw_line(coef,intercept,mi,ma)
        plt.scatter(sv[:,0], sv[:,1], color='green')
        plt.title("C = "+ str(k)+", D = ("+str(i[0])+", "+str(i[1])+")")       
        
    plt.show()
```

From the above figure. The below observations can be made for Linear SVMs.

1. When value of C is small, there is a lot of bias that is introduced to the model and hyperplane fails to linearly seperate the data for all ratios of imbalance data as the loss term in the SVM formulaiton is given less importance.

2. As the value of C increases, more importance is given to the loss term and we regularise the model by increasing the variance, thus the model performs better in seperating the classes as C increases from 0.001 to 100 even when the data is imbalanced by trying to fit the model on every point present.

3. Overall, Incase of imbalanced data, it is possible to fit a line by tuning the hyper parameter C(increase variance), However the model may not be robust to future data.







```python
#Logistic Regression C=1/lambda
regulparam = [0.001, 1, 100]
for k in regulparam:
    plt.figure(figsize=(20,5))
    for j,i in enumerate(ratios):    
        plt.subplot(1, 4, j+1)
        X_p=np.random.normal(0,0.05,size=(i[0],2))
        X_n=np.random.normal(0.13,0.02,size=(i[1],2))
        y_p=np.array([1]*i[0]).reshape(-1,1)
        y_n=np.array([0]*i[1]).reshape(-1,1)
        X=np.vstack((X_p,X_n))
        y=np.vstack((y_p,y_n))
#         scalar = StandardScaler()
#         X=scalar.fit_transform(X)
        clf = LogisticRegression(C=k, random_state=0)
        clf.fit(X, y)
        plt.scatter(X_p[:,0],X_p[:,1])
        plt.scatter(X_n[:,0],X_n[:,1],color='red')
        ma,mi= max(X[:,1]),min(X[:,1])
        intercept = clf.intercept_
        coef = clf.coef_[0]       
        draw_line(coef,intercept,mi,ma)
        plt.title("C = "+ str(k)+", D =  ("+str(i[0])+", "+str(i[1])+")")       
        
    plt.show()
```

From the above figure. The below observations can be made for Logistic Regression.

Incase logistic C= 1/lambda. 
From the formulation lambda is multiplied to the regularisation term, hence as lambda increases, it increases the bias and as it decreases, the variance increases. This interpretation is reversed as C is inverse of lambda.

1. The above convention makes the behavior of logistic regression with imbalanced data  simliar to linear SVM as C changes.
2. The formulation of logistic regression is such that, it is generally biased towards the class that has more points due to the nature of the logistic function/sigmoidal function because the hyperplane away from the datapoints has the maximum sum of signed distances.
